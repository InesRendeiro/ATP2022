```python
import TPC7

TPC7.menu()
x=TPC7.lerAlunos()
medialer=TPC7.mediaTPC(x)

function=int(input("O que pretende fazer?"))

while function != 0:
    if function == 1:
        TPC7.menu()
    if function == 2:
        print("A lista de alunos é:\n")
        print(x)
    if function == 3:
        print("Tabela de alunos e respetiva média:\n")
        lista = medialer
        for aluno in lista:
            nome = aluno["nome"]
            media = aluno["media"]
            print(f"{nome:^30}|{media:<5}")
    if function == 4:
        print("Tabela de alunos e respetivo escalão e média:\n")
        lista = TPC7.gradeTPC(medialer)
        for aluno in lista:
            nome = aluno["nome"]
            media = aluno["media"]
            grade = aluno["Escalão"]
            print(f"{nome:^30}|{media:^9}|{grade:<1}")
    if function == 5:
        TPC7.submenuDistrib()
        subfunction = int(input("Que distribuição pretende ver?"))
        while subfunction != 0:
            if subfunction == 1:
                TPC7.submenuDistrib()
            if subfunction == 2:
                print("Tabela da distribuição dos alunos por curso:\n")
                TPC7.pp(TPC7.distribCurso(x))
            if subfunction == 3:
                print("Tabela da distribuição dos alunos por escalão:\n")
                TPC7.pp(TPC7.distribEscalao(TPC7.gradeTPC(medialer)))
            subfunction = int(input("Que distribuição pretende ver agora?"))
    if function == 6:
        TPC7.submenuGraph()
        subfunction = int(input("Que gráfico pretende ver?"))
        while subfunction != 0:
            if subfunction == 1:
                TPC7.submenuGraph()
            if subfunction == 2:
                TPC7.graph(TPC7.distribCurso(x))
            if subfunction == 3:
                TPC7.graph(TPC7.distribEscalao(TPC7.gradeTPC(medialer)))
            subfunction = int(input("Que gráfico pretende ver?"))

    function = int(input("O que pretende fazer agora?"))


```
